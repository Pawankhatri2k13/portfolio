```python
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from scipy.stats import pearsonr

from CGPANN import cgpann
```

```python
input_data, output_data = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train, y_test = train_test_split(input_data, output_data.reshape(-1,1), test_size=0.2,)
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()
scaler_x.fit(X_train)
X_train = scaler_x.transform(X_train)
X_test = scaler_x.transform(X_test)
scaler_y.fit(y_train)
y_train = scaler_y.transform(y_train)
y_test = scaler_y.transform(y_test)
```

```python
num_features = X_train.shape[1]
num_outputs = y_train.shape[1]
```

```python
objCgpann = cgpann(num_features=num_features,num_outputs=num_outputs,generations=1000,
                   nodes=40,connections=4, nodesPerOut=2, outMutationChance=0.1,offsprings=10, 
                   mutationRate=0.2, learning_rate=0.2,functions=1, gene_type=3, seed=42, verbose=True)
```

```python
bestGeno = objCgpann.fit_data(X_train,y_train)
```

```python
pred = objCgpann.predict_data(X_test)
```

```python
pearsonr(pred[:,0],y_test[:,0])[0]
```
